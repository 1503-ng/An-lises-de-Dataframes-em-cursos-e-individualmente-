```
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

#Aplicando o KMeans

**Mostrando no dataset quais parâmetros serão utilizados para a clusterização e visualizando a quantidade de interações que temos.**


```

dataset = np.array([[1,4], [2,2], [2,5], [3,3], [3,5], [4,7], [5,6], [6,4], [6,7], [7,6], [7,9], [8,7], [8,9], [9,4], [9,8]])

plt.scatter(dataset[:,0], dataset[:,1])
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

```

#Definindo o número de clusters

**Definimos a utilização do algoritmo k-means, a quantidade de clusters que será utilizada 'n_clusters', o máximo de interações 'max_iter' e a quantidade que queremos na inicialização 'n_init'**

```
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300,n_init=10)
pred_y = kmeans.fit_predict(dataset)

plt.scatter(dataset[:,0], dataset[:,1])
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=70, c='red')
plt.show()
```

#Dando a cor dos dados conforme o cluster de aproximação 


```
plt.scatter(dataset[:,0], dataset[:,1], c=pred_y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=70, c='red')
plt.show()
```

#Exemplo com dados aleatórios 

```
X,y = make_blobs(n_samples=500,centers=50,random_state=0)
plt.scatter(X[:,0], X[:,1])
```

#Visualizando a curva da soma da média dos quadrados dos clusters 

**Utilizando o método de looping para fazer a relação das interações até 20 "pontos" com o WCSS**

**Quanto menor a quantidade de clusters mais interações um com os outros possuem, quanto maior o número de clusters menos relação um com o outro**

```
wcss = []

for i in range(1,20):
  kmeans= KMeans(n_clusters= i , init='k-means++', max_iter=300,n_init=10)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

#Visualizar de forma que possa ser visto a clusterização 

**Utilizando 4 clusters que seria o número ideal conforme gráfico anterior mostrando a curva**

```
kmeans= KMeans(n_clusters= 4 , init='k-means++', max_iter=300,n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=pred_y)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=50, c='red')
plt.show()
```
